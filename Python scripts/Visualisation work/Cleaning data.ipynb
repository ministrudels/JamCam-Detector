# Update all documents in collection with coordinates 

```python
# 
r = requests.get("https://api.tfl.gov.uk/Place/Type/JamCam")
response = r.json()
camera_ids = list(detection_collection.distinct('id'))
count = 0
for r in response:
    if r['id'] in camera_ids:
        geoJSON = {'coordinates': {
          'type': "Point",
          'coordinates': [r['lon'], r['lat']]
        }}
        myquery = { "id": r['id'] }
        newvalues = { "$set":  geoJSON }
        x = detection_collection.update_many(myquery, newvalues)
        print(x.modified_count, "documents updated for", r['id'])
        count+=1

print(count, 'ids updated')
```

# Create UNIX epoch from the datetime object 

```python
import datetime
import time
from datetime import timedelta

def round_to_hour(date):
    dt = datetime.datetime.strptime(date, '%Y-%m-%dT%H:%M:%S.%f')
    dt_start_of_hour = dt.replace(minute=0, second=0, microsecond=0)
    dt_half_hour = dt.replace(minute=30, second=0, microsecond=0)
    dt_end_of_hour = dt_start_of_hour + timedelta(hours=1)
    
    if dt >= dt_half_hour:
        # round up
        dt = dt_end_of_hour
    else:
        # round down
        dt = dt_start_of_hour
    return str(dt)
  
pattern = '%Y-%m-%d %H:%M:%S'

current = int(time.mktime(time.strptime(str(round_to_hour(df['new_datetime'][0])), pattern)))
t = current - 3600

df['epoch'] = df['new_datetime'].apply(lambda x: int(time.mktime(time.strptime(str(round_to_hour(x)), pattern))))
df.drop(columns=['new_datetime'], inplace=True)

df.head(n=10)
```

# Create initial object dataframes which contain all info to format into visualisation

```python
dataframe_epoch = []
for i in df['epoch'].unique():
    dataframe_epoch.append(df[df['epoch'] == i])

car = []
bus = []
truck = []
van = []
person = []

columns = []


final_df = []
# final_df = pd.DataFrame()
for i in range(len(dataframe_epoch)):
    tmp_df = dataframe_epoch[i]  

    for i in tmp_df['id'].unique():
        car.append((tmp_df['coordinates'], i, 'car', tmp_df['car'][tmp_df['id'] == i].mean(), tmp_df['epoch'].unique()[0]))
        bus.append((tmp_df['coordinates'], i, 'bus', tmp_df['bus'][tmp_df['id'] == i].mean(), tmp_df['epoch'].unique()[0]))
        truck.append((tmp_df['coordinates'], i, 'truck', tmp_df['truck'][tmp_df['id'] == i].mean(), tmp_df['epoch'].unique()[0]))
        van.append((tmp_df['coordinates'], i, 'van', tmp_df['van'][tmp_df['id'] == i].mean(), tmp_df['epoch'].unique()[0]))
        person.append((tmp_df['coordinates'], i, 'person', tmp_df['person'][tmp_df['id'] == i].mean(), tmp_df['epoch'].unique()[0]))
    
df_car = pd.DataFrame(car, columns=['coordinates', 'id','object','count','epoch'])
df_bus = pd.DataFrame(bus, columns=['coordinates', 'id','object','count','epoch'])
df_truck = pd.DataFrame(truck, columns=['coordinates', 'id','object','count','epoch'])
df_person = pd.DataFrame(person, columns=['coordinates', 'id','object','count','epoch'])

# df_car.head()
# df_bus.head()
# df_truck.head()
# df_person.head()
```

# Make the final_object_df dataframe

```python
def get_object_df(object):
    # Make final dataframe for car objects
    object_df = df[['coordinates','id']]
    object_df = object_df.drop_duplicates(["id"], keep='first').reset_index(drop=True)
    object_df.set_index('id', inplace=True, drop=True)
    object_df.insert(1, 'object', object, True)

    tmp_df = None
    if object == 'car':
        tmp_df = df_car
    if object == 'bus':
        tmp_df = df_bus
    if object == 'truck':
        tmp_df = df_truck
    if object == 'person':
        tmp_df = df_person
        
    # Merge from the epoch table
    for epoch in tmp_df['epoch'].unique():
        df2 = tmp_df.loc[tmp_df['epoch'] == epoch]
        df2 = pd.DataFrame(df2[['id', 'count']])
        df2.columns = ['id', datetime.datetime.fromtimestamp(epoch).strftime("%Y-%b-%d %H:%M")]
        object_df = pd.merge(object_df, df2, on='id', how='left')


    object_df.fillna(0)    
    object_df.head()
    return object_df
```

# Upload to mongoDB

```python
car_detection_collection.drop() 
bus_detection_collection.drop() 
truck_detection_collection.drop() 
person_detection_collection.drop() 

final_car_df = get_object_df('car')
final_bus_df = get_object_df('bus')
final_truck_df = get_object_df('truck')
final_person_df = get_object_df('person')

car_detection_collection.insert_many(final_car_df.to_dict('records'))
bus_detection_collection.insert_many(final_bus_df.to_dict('records'))
truck_detection_collection.insert_many(final_truck_df.to_dict('records'))
person_detection_collection.insert_many(final_person_df.to_dict('records'))


```
